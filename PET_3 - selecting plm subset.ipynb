# Selecting a subset for in-context learning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
pet_dataset = pd.read_csv('balanced_pet_dataset.csv')
```

```python
# code retrived from: https://stackoverflow.com/questions/23455728/scikit-learn-balanced-subsampling

balanced_sample = pet_dataset.groupby(['pet', 'adopted'], as_index=False, group_keys=False).apply(lambda s: s.sample(25, random_state=555))
```
